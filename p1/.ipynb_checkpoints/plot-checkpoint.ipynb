```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
plt.style.use("fivethirtyeight")
```

```python
result_mix  = pd.read_csv("./simout.txt", header=0, delimiter=',')
result_cpu  = pd.read_csv("./simout_cpu.txt", header=0, delimiter=',')
result_io  = pd.read_csv("./simout_io.txt", header=0, delimiter=',')


# SJF: alpha - 0.1 ~ 0.9
# SRT: alpha - 0.1 ~ 0.9
# RR: END, BEGINNNING; tslice = 10, 20, 50
# FCFS: 
```

```python
mix_avg = result_mix.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("mean")
mix_max = result_mix.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("max")
mix_min = result_mix.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("min")

```

```python
cpu_avg = result_cpu.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("mean")
cpu_max = result_cpu.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("max")
cpu_min = result_cpu.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("min")
```

```python
io_avg = result_io.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("mean")
io_max = result_io.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("max")
io_min = result_io.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("min")
```

```python
# fig.1 
# overall which is the best?

mix_avg = result_mix.loc[:, ["algo", "avg_turnaround_time"]].groupby("algo").agg("mean")
res_avg =  [mix_avg.loc["FCFS", :].avg_turnaround_time, 
             mix_avg.loc["RR", :].avg_turnaround_time,
             mix_avg.loc["SJF", :].avg_turnaround_time,
             mix_avg.loc["SRT", :].avg_turnaround_time]
res_max =  [mix_max.loc["FCFS", :].avg_turnaround_time, 
             mix_max.loc["RR", :].avg_turnaround_time,
             mix_max.loc["SJF", :].avg_turnaround_time,
             mix_max.loc["SRT", :].avg_turnaround_time]
res_min = [mix_min.loc["FCFS", :].avg_turnaround_time, 
             mix_min.loc["RR", :].avg_turnaround_time,
             mix_min.loc["SJF", :].avg_turnaround_time,
             mix_min.loc["SRT", :].avg_turnaround_time]


fig, ax = plt.subplots(1, 1)


ax.bar(range(4), res_avg, width=0.5)

# ax.errorbar(range(4), res_avg, 
#                     yerr=[[r1 - r2 for (r1, r2) in zip(res_avg, res_min)],
#                           [r1 - r2 for (r1, r2) in zip(res_max, res_avg)]], 
#                     xerr=None, 
#                     fmt='ko', ecolor='k')

#ax.errorbar(x, y, yerr=yerr, fmt='o')

ax.set_xlabel("algorithms")
ax.set_ylabel("avg. turnaround time (ms)")
ax.set_xticks(range(4))
ax.set_xticklabels(["FCFS", "RR", "SJF", "SRT"])

plt.savefig("mix_overall.png", dpi=300)
plt.show()


```
